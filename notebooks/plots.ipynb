```python
import matplotlib.pyplot as plt

true_path = '../data/results/true.out'

LFSR_path = '../data/results/LFSR.out'

LD_path = '../data/results/Halton.out'
```

```python
with open(true_path, 'r') as f:
    str = f.readline()

RMSE_true = [ float(s) for s in str.split(' ') ]

with open(LFSR_path, 'r') as f:
    str = f.readline()

RMSE_LFSR = [ float(s) for s in str.split(' ') ]

with open(LD_path, 'r') as f:
    str = f.readline()

RMSE_LD = [ float(s) for s in str.split(' ') ]
```

```python
from matplotlib.pyplot import MultipleLocator

size = len(RMSE_true)
FONTSIZE = 20
ALPHA = 0.5
LW = 2 # line width
MS = 7 # marker size
x = range(2, size + 2)

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10, 7))

ax.tick_params(axis = 'x', labelsize = FONTSIZE)
ax.tick_params(axis = 'y', labelsize = FONTSIZE)
ax.set_yscale('log')

ax.plot(x, RMSE_true, 'r*-', alpha = ALPHA, label = 'true random', linewidth = LW, markersize = MS)

ax.plot(x, RMSE_LFSR, 'b^-', alpha = ALPHA, label = 'LFSR', linewidth = LW, markersize = MS)

ax.plot(x, RMSE_LD, 'go-', alpha = ALPHA, label = 'Halton', linewidth = LW, markersize = MS)

x_major_locator = MultipleLocator(1)
ax.xaxis.set_major_locator(x_major_locator)

ax.set_xlabel('sequence length', fontsize = FONTSIZE + 10)
ax.set_ylabel('RMSE', fontsize = FONTSIZE + 10)

ax.set_title('RMSE in different sequence length w.r.t. different types of SN', fontsize = FONTSIZE)
plt.legend(fontsize = FONTSIZE)

fig.savefig('plots/RMSE_of_SNs.pdf', dpi = 300, bbox_inches = 'tight')
```
